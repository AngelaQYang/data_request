```python
import pandas as pd 
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt
import os
```

```python
INPUT_PATH = r'U:\angela\job_housing\soundcast_2050\job_housing_commute\final_set'
```

```python
def county_plot(county_id, range1, range2):
    my_county = taz_df[taz_df['county_id'] == county_id]
    my_county_df = my_county[['job_housing_ratio']]
    my_county_df.hist(bins=100, figsize=(5,5))
    plt.axis([0, range1, 0, range2]) 
    plt.show()
```

# 2014 existing

```python
year = '2014'
scenario = 'stc'
hh_tot = 'hh_' + year
TAZ_file = 'commute_taz_' + year + '_30mins_' + scenario + '.csv'
print TAZ_file
taz_df = pd.read_csv(os.path.join(INPUT_PATH, TAZ_file))
```

```python
my_df = taz_df[['mean_dist','mean_time', hh_tot,'job_housing_ratio']]
my_df.hist(bins=100, figsize=(20,15))
plt.show()
```

```python
my_df = taz_df[['job_housing_ratio']]
my_df.hist(bins=100, figsize=(10,5))
plt.axis([0,2.5, 0, 160]) 
plt.show()
```

```python
my_df = taz_df[['job_housing_ratio', hh_tot, 'county_id']]
my_df = my_df.dropna()
y = my_df[hh_tot]
plt.hist(my_df['job_housing_ratio'], bins=100, weights=y, histtype='bar', color=None)
plt.axis([0, 2.5, 0, 120000]) 
```

```python
king_df = my_df[my_df['county_id'] == 33]
king_df_x = king_df['job_housing_ratio']
king_df_y = king_df[hh_tot]
kt_df = my_df[my_df['county_id'] == 35]
kt_df_x = kt_df['job_housing_ratio']
kt_df_y = kt_df[hh_tot]
p_df = my_df[my_df['county_id'] == 53]
p_df_x = p_df['job_housing_ratio']
p_df_y = p_df[hh_tot]
s_df = my_df[my_df['county_id'] == 61]
s_df_x = s_df['job_housing_ratio']
s_df_y = s_df[hh_tot]

from matplotlib import pyplot
pyplot.hist(king_df_x, bins=100, weights=king_df_y, histtype='bar', alpha=0.5, label='king')
pyplot.hist(kt_df_x, bins=100, weights=kt_df_y, histtype='bar', alpha=0.5, label='kitsap')
pyplot.hist(p_df_x, bins=100, weights=p_df_y, histtype='bar', alpha=0.5, label='Pierce')
pyplot.hist(s_df_x, bins=100, weights=s_df_y, histtype='bar', alpha=0.5, label='Soh')
pyplot.legend(loc='upper right')
plt.axis([0, 2.5, 0, 120000]) 
pyplot.show()
```

```python
king_df = taz_df[taz_df['county_id'] == 33]
king_df_x = king_df['job_housing_ratio']
king_df_y = king_df['mean_dist']
kt_df = taz_df[taz_df['county_id'] == 35]
kt_df_x = kt_df['job_housing_ratio']
kt_df_y = kt_df['mean_dist']
p_df = taz_df[taz_df['county_id'] == 53]
p_df_x = p_df['job_housing_ratio']
p_df_y = p_df['mean_dist']
s_df = taz_df[taz_df['county_id'] == 61]
s_df_x = s_df['job_housing_ratio']
s_df_y = s_df['mean_dist']

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.scatter(king_df_x, king_df_y, s=5, c='b', marker="o", label='king')
b, m = polyfit(king_df_x, king_df_y, 1)
plt.plot(king_df_x, b + m * king_df_x, '-', c = 'b', label = 'king')

ax1.scatter(kt_df_x,kt_df_y, s=5, c='r', marker="o", label='kitsap')
b, m = polyfit(kt_df_x, kt_df_y, 1)
plt.plot(kt_df_x, b + m * kt_df_x, '-', c = 'r', label = 'kitsap')

ax1.scatter(p_df_x, p_df_y, s=5, c='g', marker="o", label='pierce')
b, m = polyfit(p_df_x, p_df_y, 1)
plt.plot(p_df_x, b + m * p_df_x, '-', c = 'g', label = 'pierce')

ax1.scatter(s_df_x,s_df_y, s=5, c='y', marker="o", label='snoh')
b, m = polyfit(s_df_x, s_df_y, 1)
plt.plot(s_df_x, b + m * s_df_x, '-', c = 'y', label = 'snoh')

plt.xlabel("mean_distance")
plt.ylabel("job_housing_ratio")
plt.legend(loc='upper right')
plt.axis([0,3.0, 0, 40]) 
plt.figure(figsize=(10,10))

plt.show()
```
