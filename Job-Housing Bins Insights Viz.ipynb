```python
import pandas as pd 
import numpy as np 
%matplotlib inline
import matplotlib.pyplot as plt
import os
```

```python
INPUT_PATH = r'U:\angela\job_housing\soundcast_2050\job_housing_commute\final_set'
```

```python
def county_plot(county_id, range1, range2):
    my_county = taz_df[taz_df['county_id'] == county_id]
    my_county_df = my_county[['job_housing_ratio']]
    my_county_df.hist(bins=100, figsize=(5,5))
    plt.axis([0, range1, 0, range2]) 
    plt.show()
```
